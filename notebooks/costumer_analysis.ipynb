# Costumer Behaviour Analysis

## 1) Imports

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## 2) Read .csv file

```python
file_path = "your_file_path_here.csv"  # Replace with your actual file path
df = pd.read_csv(file_path)
df.head()
```

## 3) Exploratory Data Analysis

```python
# Compute number of Missing Values
print("Missing Values: ")
df.isnull().sum()
```

```python
# Compute number of Duplicated Rows
print("Duplicated Rows: ")
print(df.duplicated().sum())
```

**Distributions by feature**

*Gender*

```python
df['Gender'].unique()
```

```python
# Plotting the distributions of the 'Gender' column
fig_gender, axes_gender = plt.subplots(1, 3, figsize=(18, 5))

# All data
gender_counts = df['Gender'].value_counts(normalize=True) * 100
axes_gender[0].bar(gender_counts.index, gender_counts.values, color=['pink', 'blue'])
axes_gender[0].set_title('All Data')
axes_gender[0].set_ylabel('Percentage')
axes_gender[0].set_ylim(0, 100)
for i, v in enumerate(gender_counts.values):
    axes_gender[0].text(i, v + 1, f"{v:.1f}%", ha='center')

# Purchased = 1
gender_counts_p1 = df[df['Purchased'] == 1]['Gender'].value_counts(normalize=True) * 100
axes_gender[1].bar(gender_counts_p1.index, gender_counts_p1.values, color=['pink', 'blue'])
axes_gender[1].set_title('Purchased = 1')
axes_gender[1].set_ylim(0, 100)
for i, v in enumerate(gender_counts_p1.values):
    axes_gender[1].text(i, v + 1, f"{v:.1f}%", ha='center')

# Purchased = 0
gender_counts_p0 = df[df['Purchased'] == 0]['Gender'].value_counts(normalize=True) * 100
axes_gender[2].bar(gender_counts_p0.index, gender_counts_p0.values, color=['pink', 'blue'])
axes_gender[2].set_title('Purchased = 0')
axes_gender[2].set_ylim(0, 100)
for i, v in enumerate(gender_counts_p0.values):
    axes_gender[2].text(i, v + 1, f"{v:.1f}%", ha='center')

plt.tight_layout()
plt.show()
```

*Age*

```python
# Plotting the distribution of the 'Age' column

fig, axes = plt.subplots(1, 3, figsize=(18, 5), sharey=True)

# All Data
sns.histplot(df['Age'], kde=True, color='blue', stat='density', bins=30, ax=axes[0])
axes[0].set_title('All Data')
axes[0].set_xlabel('Age')
axes[0].set_ylabel('Density')

# Purchased = 1
sns.histplot(df[df['Purchased'] == 1]['Age'], kde=True, color='green', stat='density', bins=30, ax=axes[1])
axes[1].set_title('Purchased = 1')
axes[1].set_xlabel('Age')

# Purchased = 0
sns.histplot(df[df['Purchased'] == 0]['Age'], kde=True, color='red', stat='density', bins=30, ax=axes[2])
axes[2].set_title('Purchased = 0')
axes[2].set_xlabel('Age')

plt.tight_layout()
plt.show()
```

*Estimated Salary*

```python
# Plotting the distribution of the 'Age' column

fig, axes = plt.subplots(1, 3, figsize=(18, 5), sharey=True)

# All Data
sns.histplot(df['EstimatedSalary'], kde=True, color='blue', stat='density', bins=30, ax=axes[0])
axes[0].set_title('All Data')
axes[0].set_xlabel('EstimatedSalary')
axes[0].set_ylabel('Density')

# Purchased = 1
sns.histplot(df[df['Purchased'] == 1]['EstimatedSalary'], kde=True, color='green', stat='density', bins=30, ax=axes[1])
axes[1].set_title('Purchased = 1')
axes[1].set_xlabel('EstimatedSalary')

# Purchased = 0
sns.histplot(df[df['Purchased'] == 0]['EstimatedSalary'], kde=True, color='red', stat='density', bins=30, ax=axes[2])
axes[2].set_title('Purchased = 0')
axes[2].set_xlabel('EstimatedSalary')

plt.tight_layout()
plt.show()
```
